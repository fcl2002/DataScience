```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
#variável em uma distribuição normal com a função rvs (100 elementos)
dados = norm.rvs(size = 1000)
dados
```

```python
#histograma
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
#gráfico para verificar se a dstribuição é normal
fig, ax = plt.subplots()
stats.probplot(dados, fit = True, plot = ax)
plt.show()
```

```python
#teste de Shapiro
stats.shapiro(dados)
```

```python
#Dados não normais
dados2 = skewnorm.rvs(4, size = 1000)
```

```python
#histograma 2
plt.hist(dados2, bins = 20)
plt.title('Dados 2')
```

```python
#visualização da distribuição
fig, ax = plt.subplots()
stats.probplot(dados2, fit = True, plot = ax)
plt.show()
```

```python
stats.shapiro(dados2)
#resultado bem menor que 0.005, logo notamos que segundo a hipótese de Shapiro temos uma distribuição não normal
```

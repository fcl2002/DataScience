```python
import pandas as pd
import matplotlib.pyplot as plt

base = pd.read_csv('trees.csv')
print("shape:", base.shape)
base.head()
```

```python
#imprimindo juntos
#criação de figuras,na qual os gráficos serão posicionados
plt.figure(1) #cria objeto 1
plt.subplot(2, 2, 1) #n° linhas, n° colunas e a posição que irá ser impressa
plt.scatter(base.Girth, base.Volume)
plt.subplot(2, 2, 2)
plt.scatter(base.Girth, base.Height)
plt.subplot(2, 2, 3)
plt.scatter(base.Height, base.Volume, marker = '*')
plt.subplot(2, 2, 4)
plt.hist(base.Volume)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

base = pd.read_csv('trees.csv')
print("shape:", base.shape)
base.head()
```

```python
#patch_artist = True preenche, showfliers outliers
#pode comparar varias variáveis contínuas lado a lado
#vert = false (horizontal), showfliers (mostrar outliers), notch (mostrar entalhe), patch (preenchimento)
plt.boxplot(base.Volume, vert = False, showfliers = False, notch = True, patch_artist = True)
plt.title('Árvores')
plt.xlabel('Volume')
```

```python
#dados por coluna
plt.boxplot(base)
plt.title('Árvores')
plt.xlabel('Dados')
```

```python
#Geração de 3 boxplots, cada um mostrando diferentes informações
plt.boxplot(base.Volume, vert = False)
plt.boxplot(base.Girth, vert = False)
plt.boxplot(base.Height, vert = False)
plt.title('Árvores')
plt.xlabel('Dados')
```

Histograma e Densidade com Seaborn

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

base = pd.read_csv('trees.csv')
base.head()
```

```python
#histograma com 10 divisões(bins) e com gráfico de densidade (kernel density - suavização)
#stat: count,frequency, probability, percent, density
sb.histplot(base.Volume, color = 'blue', kde = True, stat = "density", bins = 10).set(title = 'Volume')
```

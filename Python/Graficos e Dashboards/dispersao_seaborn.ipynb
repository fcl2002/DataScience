```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

base = pd.read_csv('co2.csv')
base
```

```python
#gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type
sb.scatterplot(x = base.conc, y = base.uptake, hue = base.Type)
```

```python
#seleção de registros específicos da base de dados (Quebec e Mississipi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
#subgráfico (1 linha x 2 colunas) mostrando gráficos sobre cada região
plt.figure()
plt.subplot(1, 2, 1)
sb.scatterplot(x = q.conc, y = q.uptake).set_title('Quebec')
plt.subplot(1, 2, 2)
sb.scatterplot(x = m.conc, y = m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
#refrigerado e não refrigerado
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
#gráfico somente com 'chilled' e 'nonchilled'
plt.figure()
plt.subplot(1, 2, 1)
sb.scatterplot(x = ch.conc, y = ch.uptake).set_title('Chilled')
plt.subplot(1, 2, 2)
sb.scatterplot(x = nc.conc, y = nc.uptake).set_title('Non Chilled')
plt.tight_layout()
```
